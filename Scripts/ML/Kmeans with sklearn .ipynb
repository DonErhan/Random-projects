## *Generate  data*

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_blobs(n_samples=1000000, n_features=2, centers=5, random_state=42, cluster_std=1.0)
```

```python
plt.scatter(X[:,0], X[:,1],c=y)
plt.title('Dataset ploting')
plt.show()
```

```python
O = {0: [-6.88397533693082, -6.881152237145192],
 1: [-2.507390888195659, 9.012348535834398],
 2: [4.666490257528011, 1.9482277745687282],
 3: [-8.841685411707159, 7.3227460048848725],
 4: [1.9956809274535505, 4.185773373141004]}
```

```python
new = np.asarray([O[i] for i in O],np.float64)
```

```python
%%time
kmeans = KMeans(n_clusters=5,max_iter=800,init=new,n_init=1,tol=0.001)
kmeans.fit(X)
y_km = kmeans.fit_predict(X)
clusters= kmeans.cluster_centers_
```

```python
# identifi centroids
clusters= kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0], X[:,1], c=y_km)
plt.scatter(clusters[:,0], clusters[:,1],marker='$c$',c='red')
plt.title('Clusters with centroids')
plt.show()
```
